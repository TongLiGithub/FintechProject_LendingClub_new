## 2. Format Conversion of Feature Names 

Feature names in the current data have both uppercase and lowercase, whereas feature names in the historic data have underscores. So I'm going to convert the feature names in the two datasets into the same format, and then find out the common features and the unique features in the two datasets.

```python
# Convert feature names in the current data
current_feature = list(df_current.columns.values)
for i in range (len(current_feature)):
    current_feature[i] = str.lower(str(current_feature[i]))
current_feature[:5]
```

```python
# Convert feature names in 2014 data
his_feature = list(df_2014.columns.values)
for i in range (len(his_feature)):
    his_feature[i] = str(his_feature[i]).replace('_','')
his_feature[:5]
```

## 3. Find Common Features in the Two Datasets and Save for 2014 Loans

```python
# find out and print common features of the two datasets
common_feature = set(his_feature)&(set(current_feature))
print (len(common_feature))
print (common_feature)
```

```python
his_unique=set(his_feature)-set(current_feature)
current_unique=set(current_feature)-set(his_feature)
print (len(current_unique), current_unique)
print (len(his_unique), his_unique)
```

```python
feature_dic={}
feature_dic['common_feature']=common_feature
feature_dic['his_unique']=his_unique
feature_dic['current_unique']=current_unique
```

```python
# Add common features that have different names for current and 2014 data. Use the form in the 2014 data
common_feature=common_feature.union({'zipcode', 
 'fundedamnt',
 'verificationstatus',
 'verificationstatusjoint',
 'loanamnt',
 'numacctsever120pd'})
len(common_feature)
```
